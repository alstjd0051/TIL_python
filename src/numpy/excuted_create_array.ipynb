# NumPy arange와 Inspace 데이터 생성
- np.arrage: N만큼 차이 나는 숫자 생성
- np.linspace: N 등분한 숫자 생성


## np.arage와 np.linspace 차이점
- np.arage([start],stop,[step]..)
    - [] 생략가능, 끝 값 포함안함
    - 장점: step, 범위를 구간, 간격 강조할때 사용하면 코드 가독성 UP
- np.linspace(start,stop,num=50..)
    - 처음 값과 끝 값 포험, 몇개로 만들지 매개변수로 줌
    - 장점: 개수 강조할때 사용하면 코드 가독성 UP

```python
# 연속적인 데이터 생성
# - np.arage(): N 만큼 차이나는 숫자 생성
# - np.linspace(): N 등분한 숫자생성
import numpy as np
print(np.arange(10,30,5))
```

```python
# 0이상 2미만 까지 0.3씩 차이나게 생성

print(np.arange(0,2,0.3))
```

```python
# 0~99까지 100등분
x =- np.linspace(0,99,100)
print(x)
```

```python
# arange 함수는 끝 값ㄱ을 포함하지 않기 때문에
# 가독성을 위해서 1.25를 1+0.25로 표현
# 0부터 1.25 미만까지(끝 값 포함 안하니까 1까지 출력) 0.25씩 차이나게 생성
print(np.arange(0, 1+0.25, 0.25))
```

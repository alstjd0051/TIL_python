```python
# 입력값으로 나이를 받아서, 60세가 넘으면 김치찌개를 추천하고, 나머지 사람들에게는 볶음밥을 추천하는 음식추천 함수 rec_food 개발
```

```python
def rec_food(age: int) -> str:
    if age > 60:
        return "김치찌개"
    else:
        return "볶음밥"
```

```python
rec_food(20)
```

```python
from urllib.request import urlopen
# 1. 사이트접속하기
url = "https://ai-dev.tistory.com/1"
html = urlopen(url)
response = html.read()
# 2. html 문서 가져오기
bs_obj = BeautifulSoup(response,"html.parser")
# 3. beautifulsoup 객체로 저장
bs_obj.find_all("p")

# 4. 내가 원하는 태그 속 테스트 가져오기
text = bs_obj.find_all("p")
text[0].text
```

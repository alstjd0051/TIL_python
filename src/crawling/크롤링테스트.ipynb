```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
import time

# Chrome 옵션 설정
chrome_options = Options()
# headless 모드 (브라우저 창을 띄우지 않음) - 필요시 주석 해제
# chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')

# 웹드라이버 초기화
driver = webdriver.Chrome(options=chrome_options)
url = "https://www.naver.com/"

try:
    driver.get(url)
    # 페이지가 완전히 로드될 때까지 대기
    time.sleep(3)
finally:
    pass
```

```python
# ContentHeaderView-module__header_tab___WW6fU 클래스를 가진 요소 찾기
try:
    # 요소가 로드될 때까지 대기 (최대 10초)
    header_tab = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.CLASS_NAME, "ContentHeaderView-module__header_tab___WW6fU"))
    )
    
    # 텍스트 데이터 추출
    tab_text = header_tab.text.strip()
    print(f"추출된 텍스트: {tab_text}")
    
    # 개별 탭 항목으로 분리 (공백 기준)
    tab_items = tab_text.split()
    print(f"\n개별 탭 항목:")
    for i, item in enumerate(tab_items, 1):
        print(f"{i}. {item}")
        
except Exception as e:
    print(f"요소를 찾을 수 없습니다: {e}")
```

```python
# CSS 선택자를 사용한 대안 방법
# DOM Path를 기반으로 한 선택
try:
    css_selector = "div#wrap > div#container > div#root > div.Layout-module__column_left___tLO23 > div#newsstand > div.ContentHeaderView-module__content_header___nSgPg > div.ContentHeaderView-module__header_tab___WW6fU"
    
    header_tab_css = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.CSS_SELECTOR, css_selector))
    )
    
    tab_text_css = header_tab_css.text.strip()
    
    # 개별 탭 항목으로 분리
    tab_items_css = tab_text_css.split()
    for i, item in enumerate(tab_items_css, 1):
        print(item)
        
except Exception as e:
    print(f"CSS 선택자로 요소를 찾을 수 없습니다: {e}")
```

```python
# 웹드라이버 종료
driver.quit()
```

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup

url = "https://www.naver.com/"
html = urlopen(url)
bs_obj = BeautifulSoup(html, "html.parser")

# CSS 선택자로 헤더 탭 요소 찾기 (querySelector 스타일)
header_tab = bs_obj.select_one("div.ContentHeaderView-module__header_tab___WW6fU")

if header_tab:
    # 텍스트 내용 추출
    content = header_tab.get_text(strip=True)
    print(f"헤더 탭 내용: {content}")
    
    # 개별 탭 항목으로 분리 (공백 기준)
    tabs = content.split()
    print(f"\n탭 항목들:")
    for idx, tab in enumerate(tabs, 1):
        print(f"  {idx}. {tab}")
else:
    print("헤더 탭 요소를 찾을 수 없습니다.")
    
    # 대안 선택자 시도
    print("\n대안 선택자 시도 중...")
    alternatives = [
        "div.ContentHeaderView-module__content_header___nSgPg > div.ContentHeaderView-module__header_tab___WW6fU",
        "div#newsstand div.ContentHeaderView-module__header_tab___WW6fU",
        "div[class*='header_tab']"
    ]
    
    for selector in alternatives:
        result = bs_obj.select_one(selector)
        if result:
            print(f"선택자 '{selector}'로 요소를 찾았습니다.")
            print(f"내용: {result.get_text(strip=True)}")
            break
```

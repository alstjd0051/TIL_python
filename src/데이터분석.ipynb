```python
# 상속
class Country01:
    name = '국가명'
    capital = '수도'
    population = 100
    
    def introduce(self):
        sentence = "국가소개 문장입니다."
        return sentence

Country01.name
```

```python
class Korea01(Country01):
    def __init__(self, name):
        self.name = name

    def explanation(self):
        sentence = "국가 이름은 {}입니다.".format(self.name)
        return sentence
```

```python
ex3 = Korea01(name="한국")
ex3.name
ex3.capital
ex3.population
ex3.introduce()
ex3.explanation()
```

```python
# 메소드 오버라이딩
class Korea02(Country01):
    def __init__(self, name):
        self.name = name

    def explanation(self):
        sentence = "국가 이름은 {}입니다.".format(self.name)
        return sentence
```

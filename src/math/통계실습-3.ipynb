```python
# 동전 여러번 던져보기
n: int = 10 # 동전 던지는 횟수
heads: int = 0 # 앞면 개수 저장

for i in range(n):
    if random.choice(["H", "T"]) == "H":
        heads += 1
print(f"총 던진 횟수 {n}")
print(f"앞면이 나온 횟수 {heads}")
print(f"앞면이 나온 비율 {heads/n}")
```

```python
# 던질수록 앞면 비율이 어떻게 변하는지 기록하기
from typing import List


n:int = 1000 # 동전 던지는 횟수
heads:int = 0 # 앞면 개수 저장
history:List[int] = [] # 앞면 비율 기록용 리스트

for i in range(1,n+1):
    if random.choice(["H", "T"]) == "H":
        heads +=1 
    history.append(heads/i) # 현재까지의 앞면 비율 저장

print(f"마지막 앞면 비율: {history[-1]}") # 마지막 앞면 비율 출력
```

```python
# 이항분포 PMF 계산하기
import matplotlib.pyplot as plt
import math
plt.rcParams['font.family'] = 'NanumGothic'
plt.rcParams['axes.unicode_minus'] = False

def binom_pmf(n,p,k): # 이항분포 확률질량함수(PMF)를 계산하는 함수정의(입력: n,p,k)
    comb = math.comb(n,k) # n번중k번 성공하는 경우의 수를 계산
    return comb * (p**k) * ((1-p)**(n-k)) # nCk * p^k * (1-p)^(n-k) 공식 계산 후 반환

n = 10 # n=시행횟수
p= 0.3 # p=성공확률(0.3)
xs = list(range(n+1)) # k는 0부터 n까지 가능하므로 리스트 새엇ㅇ
ps = [binom_pmf(n,p,k) for k in xs] # 각 k에 대해서 확률 P(X=k)를 계산해 리스트로 저장

print(list(zip(xs,ps))) # (k,확률) 형태로 묶어서 한번에 보기 좋게 출력
plt.bar(xs,ps)
total_prob = sum(ps) # 모든 k에 대한 확률을 더하면 1에 가까움
print(f"확률의 총합: {total_prob}") # 확률의 총합을 출력
```

```python
# 이항분포 PMF 계산하기
import matplotlib.pyplot as plt
import math
plt.rcParams['font.family'] = 'NanumGothic'
plt.rcParams['axes.unicode_minus'] = False

def binom_pmf(n,p,k): # 이항분포 확률질량함수(PMF)를 계산하는 함수정의(입력: n,p,k)
    comb = math.comb(n,k) # n번중k번 성공하는 경우의 수를 계산
    return comb * (p**k) * ((1-p)**(n-k)) # nCk * p^k * (1-p)^(n-k) 공식 계산 후 반환

n = 10 # n=시행횟수
p= 0.3 # p=성공확률(0.3)
xs = list(range(n+1)) # k는 0부터 n까지 가능하므로 리스트 새엇ㅇ
ps = [binom_pmf(n,p,k) for k in xs] # 각 k에 대해서 확률 P(X=k)를 계산해 리스트로 저장

print(list(zip(xs,ps))) # (k,확률) 형태로 묶어서 한번에 보기 좋게 출력
plt.bar(xs,ps)
total_prob = sum(ps) # 모든 k에 대한 확률을 더하면 1에 가까움
print(f"확률의 총합: {total_prob}") # 확률의 총합을 출력
```

```python
# zip(xs, ps): xs와 ps 리스트를 묶어서 (k, prob) 형태의 쌍으로 만들어줌
# 즉, 각 k마다 그에 해당하는 확률 prob를 동시에 순회할 수 있게 해준다
for k, prob in zip(xs, ps): # xs(0~n)와 ps(확률)을 동시에 꺼내서 반복
    # 각 k(성공 횟수)에 대한 이항분포 확률 pmf 값을 보기 좋게 출력
    # k: 성공 횟수, prob: 그때의 확률
    print(f"k={k:2d}일 확률 P(X={k}) = {prob:.3f}")
```

```python
import numpy as np

np.random.seed(42)

mu = 170
sigma = 6
n = 1000

heights = np.random.normal(
    loc=mu,
    scale= sigma,
    size=n
)
```

```python
#데이터가 잘 만들어졌는지 확인
print(heights[:10])
print(f"데이터개수: {len(heights)}")
```

```python
# 요약 통계 직접 구하기
mean_value = np.mean(heights)
std_value = np.std(heights)
min_value = np.min(heights)
max_value = np.max(heights)

print(f"평균: {mean_value}")
print(f"표준편차: {std_value}")
print(f"최소: {min_value}")
print(f"최대: {max_value}")
```

```python
# 히스토그램으로 분포 확인하기
import matplotlib.pyplot as plt
plt.hist(heights,bins=30) # height를 30개 구간으로 나누어 히스토그램 그리기
plt.axvline(mu)
plt.axvline(mu-sigma)
plt.axvline(mu+sigma)
plt.xlabel('Height (cm)')
plt.ylabel('Count (num)')
plt.title("Mean and +- 1∑") # 그래프 제목
plt.show()

lower = mu- sigma
upper = mu + sigma
inside = (heights >= lower) & (heights <= upper)
ratio = inside.mean()

print(f"mean±1∑ inner rate: {ratio}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm # 정규분포와 확률밀도함수(pdf) 등을 쓰기 위한 도구

np.random.seed(42)

mu = 170 # 평균
sigma = 6 # 표준편차
n = 1000 # 표본 개수

heights = np.random.normal(
    loc=mu,
    scale=sigma,
    size=n
) # 정규 분포에서 키 데이터 n개를 생성

plt.figure(figsize=(10,5)) # 그래프 크기를 가로 10 세로 5로 설정
plt.hist(heights,bins=30,density=True,alpha=0.5) # 히스토그램 그리기 density=True로 확률밀도형태 맞춤
#bins=30: 막대를 30개 구간으로 나눔, density=True -> 막대의 면적이 1이 되도록 스케일 조정, alpha=0.5 -> 히스토그램 막대 반투명하게 보여줌

xs = np.linspace(150, 190, 200) # x축에 사용할 값들을 150~190 사이에서 200개 만듦

# pdf는 probability density function(확률밀도함수)의 약자입니다.
# 정규분포의 pdf는 특정 값에서의 확률 '밀도'를 의미하며, 이 함수의 전체 면적은 1입니다.
# 각각의 xs 값에 대해, 해당 위치에서의 확률밀도(pdf) 값을 계산해서 곡선을 그릴 수 있습니다.
pdf = norm.pdf(xs, loc=mu, scale=sigma) # 정규분포의 pdf의 값을 xs 위치에서 계산

plt.plot(xs, pdf, linewidth=2) # xs(가로축)과 pdf(세로축)을 연결해 정규곡선을 그림
plt.title("Histogram + Normal Curve (PDF)")
plt.xlabel("Height (cm)")
plt.ylabel("Density (probability)")
plt.show()
```

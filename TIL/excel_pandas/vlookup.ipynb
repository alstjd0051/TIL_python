# pandas merge


> 데이터 프레임들을 database에서와 같은 병합을 하게 해주는 함수
>
> 엑셀의 vlookup같은 병합도 가능하다.

```python
import pandas as pd
from pathlib import Path
from icecream import ic
```

```python
pd.set_option('display.max_rows', 6) # 최대 6줄까지 출력
url = Path("assets") / "02_02_merge01.xlsx"
```

```python
df1 = pd.read_excel(url)
df1
```

```python
df2 = pd.read_excel(url, sheet_name=1)
df2
```

```python
df3 = df1.merge(df2, how='left', on='제품')
df3.to_excel(Path("assets") / "merge_result01.xlsx")
```

### 1-1. SVD êµ¬í˜„ì„ ìœ„í•œ ì´ë¡

#### ğŸ“ ì„¤ëª…: SVD (Singular Vector Decomposition) ë€?
íŠ¹ì´ê°’ í–‰ë ¬ë¶„í•´, SVDë€ í–‰ë ¬ $M \in{\mathbb{R}^{m \times n}}$ë¥¼ ë‹¤ìŒê³¼ ê°™ì´ íŠ¹ë³„í•œ ì„±ì§ˆì„
ê°€ì§„ 3ê°œì˜ í–‰ë ¬ë“¤ì˜ ê³±ìœ¼ë¡œ ë‚˜íƒ€ë‚´ëŠ” ê³¼ì •ì„ ë§í•©ë‹ˆë‹¤.

$M=U\Sigma V^T$.

ì—¬ê¸°ì„œ 3ê°œì˜ í–‰ë ¬ U, $\Sigma$, VëŠ” ê°ê°
1. $U$ëŠ” mì°¨ì› ì •ê·œ ì§êµ í–‰ë ¬ (orthonormal matrix)
2. $\Sigma$(sigma)ëŠ” singular valueë¥¼ ì„±ë¶„ìœ¼ë¡œ í•˜ëŠ” ëŒ€ê° í–‰ë ¬(diagonal matrix)
3. $V$ëŠ” nì°¨ì› ì •ê·œ ì§êµ í–‰ë ¬ (orthonormal matrix)

ì´ë¼ëŠ” íŠ¹ì„±ì´ ìˆìŠµë‹ˆë‹¤.



### 1-2. SVD êµ¬í˜„ì„ ìœ„í•œ ë³´ì¡°í•¨ìˆ˜ ì •ì˜
> ë³¸ê²©ì ì¸ ë‚´ìš© ì„¤ëª…ì— ì•ì„œ, í–‰ë ¬ì€ ê·¸ í¬ê¸°ê°€ ì»¤ì§ˆìˆ˜ë¡ ë‹¨ìˆœíˆ printí–ˆì„ ë•Œ ë³´ê¸° ë§¤ìš° ë¶ˆí¸í•´ì§€ê¸° ë•Œë¬¸ì— ì‹œê°í™”ë¥¼ ìœ„í•œ ë³´ì¡°í•¨ìˆ˜ë¥¼ ì •ì˜í•´ë´…ë‹ˆë‹¤.


#### ğŸ‘¨â€ğŸ’» ì½”ë“œ : í–‰ë ¬ì˜ ì‹œê°í™”ë¥¼ ìœ„í•œ ë³´ì¡°í•¨ìˆ˜
- í–‰ë ¬ ì‹œê°í™”ë¥¼ ìœ„í•œ ë³´ì¡°í•¨ìˆ˜ `plot_matrix` ì •ì˜
- ëœë¤ ë°ì´í„°í–‰ë ¬ì„ ìƒì„±í•œ ë’¤ `plot_matrix`í•¨ìˆ˜ë¥¼ í†µí•´ ì‹œê°í™”

````python
# í–‰ë ¬ ì‹œê°í™”ë¥¼ ìœ„í•œ ë³´ì¡°í•¨ìˆ˜ ì •ì˜
def plot_matrix(matrix, numbers=True, size_scale=0.7):
    """
    matrixë¡œ ì£¼ì–´ì§„ í–‰ë ¬ì„, ë³´ê¸°ì‰½ê²Œ ê·¸ë ¤ì£¼ëŠ” í•¨ìˆ˜ì…ë‹ˆë‹¤.
    í–‰ë ¬ì˜ í¬ê¸°ê°€ í° ê²½ìš° numbersë¥¼ Falseë¡œ ì„¤ì •í•´ ê° ì›ì†Œì˜ ê°’ì´ í‘œì‹œë˜ì§€ ì•Šë„ë¡ í•˜ê³ ,
    size_scaleì„ ì¤„ì—¬ ì „ì²´ í¬ê¸°ë¥¼ ì¡°ì ˆí•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.
    """
    n_rows, n_cols = matrix.shape
    # í–‰ë ¬ í¬ê¸°ì— ë¹„ë¡€í•˜ë„ë¡ figureì˜ ì‚¬ì´ì¦ˆ ì„¤ì •
    figure_size = (size_scale * n_cols, size_scale * n_rows)
    fig, ax = plt.subplots(figsize=figure_size)
    # ë¶ˆí•„ìš”í•œ ë¶€ë¶„ë“¤ ë¹„í™œì„±í™”
    viz_args = dict(cmap='Purples', cbar=False, xticklabels=False, yticklabels=False)
    sns.heatmap(data=matrix, annot=numbers, fmt='.2f', linewidths=.5, **viz_args)
````

````python
# ì‹¤ìŠµíŒŒì¼ì„ ì—¬ëŸ¬ë²ˆ ì‹¤í–‰í•´ë„ ê°™ì€ ê²°ê³¼ê°€ ë‚˜ì˜¤ë„ë¡ random seedë¥¼ ê³ ì •í•©ë‹ˆë‹¤.
np.random.seed(1234)

# 6x9 í¬ê¸°ì˜ ëœë¤ í–‰ë ¬ì„ ìƒì„±í•©ë‹ˆë‹¤.
M = np.random.randn(6, 9)

# ë‹¨ìˆœ printì„ ì´ìš©í•˜ì—¬ í–‰ë ¬ ì‹œê°í™”
print(M)
````

````python
# ë³´ì¡°í•¨ìˆ˜ë¥¼ í™œìš©í•˜ì—¬ í–‰ë ¬ ì‹œê°í™”
plot_matrix(M)
````

### 1-3. SVD êµ¬í˜„ ì‹¤ìŠµ

> SVDë¥¼ êµ¬í˜„í•©ë‹ˆë‹¤.

#### ğŸ“ ì„¤ëª…: ì „ì²´ SVDê³¼ì •ì„ í™•ì¸í•˜ê¸° ìœ„í•œ ë³´ì¡°í•¨ìˆ˜ ì •ì˜

SVDëŠ” numpy ë¼ì´ë¸ŒëŸ¬ë¦¬ë¥¼ í™œìš©í•˜ì—¬ ìˆ˜í–‰í•˜ëŠ”ë°,
numpyì—ì„œ SVDë¥¼ ìˆ˜í–‰í•´ì£¼ëŠ” ì´ í•¨ìˆ˜(`np.linalg.svd`)ëŠ” ëŒ€ê°í–‰ë ¬ sigmaì˜ 0ì¸ ë¶€ë¶„ì„ ìƒëµí•˜ê³  ëŒ€ê°ì„±ë¶„ì¸ singular valueë“¤ë§Œ arrayí˜•íƒœë¡œ ë¦¬í„´í•´ì£¼ë„ë¡ êµ¬í˜„ë˜ì–´ìˆìŠµë‹ˆë‹¤.

ì‹¤ì œë¡œ ì‚¬ìš©í• ë•ŒëŠ” í›¨ì”¬ ì‹¤ìš©ì ì¸ ë°©ì‹ì´ê¸°ëŠ” í•˜ì§€ë§Œ, ê°•ì˜ì—ì„œëŠ” ì´ë¡ ìƒ ì •í™•í•œ ì„¤ëª…ì„ ìœ„í•´
ì™„ì „í•œ ëŒ€ê°í–‰ë ¬ì„ ë°˜í™˜í•´ì£¼ëŠ” ìƒˆë¡œìš´ í•¨ìˆ˜ `full_svd`ë¥¼ ì •ì˜í•´
ì‚¬ìš©í•˜ê² ìŠµë‹ˆë‹¤.


#### ğŸ‘¨â€ğŸ’» ì½”ë“œ : ë³´ì¡°í•¨ìˆ˜ ì •ì˜
- ì „ì²´ SVDê³¼ì • í™•ì¸ì„ ìœ„í•œ ë³´ì¡°í•¨ìˆ˜ `full_svd` ì •ì˜
- ì´ë¥¼ ì´ìš©í•´ ëœë¤ í–‰ë ¬ì— svdì ìš©

````python
def full_svd(matrix):
    # numpyë¥¼ ì´ìš©í•œ SVDë¥¼ ìˆ˜í–‰í•©ë‹ˆë‹¤.
    U, singular_values, V = np.linalg.svd(matrix)

    # numpyì˜ svd ê²°ê³¼ë¡œ ë‚˜ì˜¤ëŠ” sigmaì˜ diagonal ì„±ë¶„ì„ ê°€ì§€ê³  diagonal matrixë¥¼ ë³µì›í•´ì¤ë‹ˆë‹¤.
    m, n = matrix.shape       # matrix í–‰ë ¬ì˜ ì°¨ì›
    sigma = np.zeros([m, n])  # matrix í–‰ë ¬ê³¼ ê°™ì€ ì°¨ì›ì˜ ì˜í–‰ë ¬ì„ ë§Œë“¤ì–´ë‘¡ë‹ˆë‹¤.

    rank = len(singular_values)  # rank ê³„ì‚°
    sigma[:rank, :rank] = np.diag(singular_values)  # rankê¹Œì§€ë§Œ ë³µì›
    return U, sigma, V.T
````

````python
# SVD ìˆ˜í–‰
U, Sigma, V = full_svd(M)
````


#### ğŸ‘¨â€ğŸ’» ì½”ë“œ : SVDê²°ê³¼ë¡œë¶€í„° ì›ë³¸ í–‰ë ¬ ë³µì›
- ë¶„í•´ ê²°ê³¼ í–‰ë ¬ì„ í†µí•´ ì›ë³¸ í–‰ë ¬ì´ ë³µì›ë˜ëŠ”ì§€ í™•ì¸

````python
# íŒŒì´ì¬ì˜ @ ì—°ì‚°ìë¥¼ ì´ìš©í•˜ë©´ np.dot í•¨ìˆ˜ì™€ ê°™ì€ í–‰ë ¬ê³±ì´ë‚˜ ë‚´ì ì„ í¸ë¦¬í•˜ê²Œ í˜¸ì¶œí•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.
restored = U @ Sigma @ V.T # TëŠ” ì „ì¹˜í–‰ë ¬(transpose)

# í–‰ë ¬ ì‹œê°í™”
plot_matrix(restored)
print("Maximum diff: ", np.abs(M - restored).max())
````

í–‰ë ¬ U, $\Sigma$, VëŠ” ê°ê° ë‹¤ìŒê³¼ê°™ì€ shapeì„ ê°€ì§€ê³  ìˆì–´, ìˆœì„œëŒ€ë¡œ ê³±í–ˆì„ ë•Œ ì›ë˜ì˜ í–‰ë ¬ê³¼ ê°™ì€ í¬ê¸°ê°€ ë©ë‹ˆë‹¤.

#### ğŸ“ ì„¤ëª… : ì •ê·œì§êµí–‰ë ¬(orthonormal matrix) U, V
í–‰ë ¬ Uì™€ VëŠ” ëª¨ë‘ ì„ í˜•ëŒ€ìˆ˜í•™ì—ì„œ ì •ê·œì§êµí–‰ë ¬(orthonormal matrix)ì´ë¼ê³  ë¶€ë¥´ëŠ” í–‰ë ¬ì…ë‹ˆë‹¤.
ì´ í–‰ë ¬ì€ ì‹œê°í™” ê²°ê³¼ì—ì„œ ë³´ì´ë“¯ì´ ìœ¡ì•ˆìœ¼ë¡œ ëˆˆì— ë„ëŠ” íŠ¹ì„±ì´ ìˆëŠ”ê²ƒì€ ì•„ë‹ˆì§€ë§Œ, ê° í–‰ì„ ë³„ê°œì˜ ë²¡í„°ë¡œ ë³¼ë•Œ ëª¨ë‘ê°€ ì„œë¡œ ì§êµí•˜ê³ , ê¸¸ì´(norm)ê°€ 1ì¸ í–‰ë ¬ë¡œ ì •ì˜ë©ë‹ˆë‹¤.

ì´ëŸ° ì •ì˜ë¡œ ì¸í•´ ì •ê·œì§êµí–‰ë ¬ì€ ìê¸°ìì‹ ì˜ transposeì™€ ê³±í–ˆì„ ë•Œ ë‹¨ìœ„í–‰ë ¬ì´ ë‚˜ì˜¨ë‹¤ëŠ”, ì¦‰
ë‹¨ìˆœíˆ transposeë§Œ í•´ì£¼ë©´ ì—­í–‰ë ¬ì´ ëœë‹¤ëŠ” í¸ë¦¬í•œ íŠ¹ì„±ì„ ì§€ë‹™ë‹ˆë‹¤.


#### ğŸ‘¨â€ğŸ’» ì½”ë“œ : U, Vì˜ ì •ê·œì§êµí–‰ë ¬ë¡œì„œì˜ íŠ¹ì„± í™•ì¸
- ì •ê·œì§êµí–‰ë ¬ì˜ ì •ì˜ì— ë”°ë¥¸ íŠ¹ì„± í™•ì¸
- ì—­í–‰ë ¬ê³¼ ê³±í•´ì„œ ë‹¨ìœ„í–‰ë ¬ì´ ë‚˜ì˜¤ëŠ”ê²ƒ í™•ì¸

ë‹¤ìŒ ë‘ ì…€ì€ Vì— ëŒ€í•´ ìˆ˜í–‰í•´ë„, í–‰ë ¬ì˜ í¬ê¸°ë§Œ ë‹¤ë¥¼ ë¿ ê°™ì€ ê²°ê³¼ë¥¼ í™•ì¸í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.

ì´ì œ, $\Sigma$ì˜ íŠ¹ì„±ì„ í™•ì¸í•´ë³´ê² ìŠµë‹ˆë‹¤.

````python
# Sigma í–‰ë ¬ ì‹œê°í™”
plot_matrix(Sigma)
````

$\Sigma$ëŠ” singular valueë¥¼ ëŒ€ê°ì„±ë¶„ìœ¼ë¡œ í•˜ëŠ” ëŒ€ê°í–‰ë ¬ì´ë©°, SVDì˜ ê²°ê³¼ì—ì„œ ì´ ëŒ€ê°ì„  ì„±ë¶„ì€
ë³´ì‹œë‹¤ì‹œí”¼ ë’¤ë¡œ ê°ˆìˆ˜ë¡ ê°’ì´ ì‘ì•„ì§€ë„ë¡ ì •ë ¬ë˜ì–´ìˆìŠµë‹ˆë‹¤. singular valueëŠ” í–‰ë ¬ì˜ eigen valueì™€
ê´€ë ¨ëœ ê°œë…ìœ¼ë¡œ, ì´ì™€ ê´€ë ¨ë˜ì–´ì„œ SVDì˜ ìˆ˜ë§ì€ í¥ë¯¸ë¡œìš´ íŠ¹ì„±ë“¤ì„ ìœ ë„í•´ë‚¼ ìˆ˜ ìˆì§€ë§Œ, ì—¬ê¸°ì—ì„œëŠ” ìƒëµí•˜ê³ , ë‹¤ìŒì˜ íŠ¹ì„±ì— ì£¼ëª©í•´ë³´ê² ìŠµë‹ˆë‹¤.

í–‰ë ¬ $\Sigma$ì—ì„œ ë‹¨ í•œê°œ, kë²ˆì§¸ì˜ ê°’ë§Œ ë‚¨ê¸°ê³  ë‚˜ë¨¸ì§€ë¥¼ ì „ë¶€ 0ìœ¼ë¡œ ë§Œë“  $\Sigma_k$ë¼ëŠ” í–‰ë ¬ì„ ìƒê°í•´ë³´ê² ìŠµë‹ˆë‹¤.

````python
# í–‰ë ¬ì˜ (k,k)ë²ˆì§¸ ì›ì†Œë§Œ ë‚¨ê¸°ê³  ë‚˜ë¨¸ì§€ë¥¼ 0ìœ¼ë¡œ ë§Œë“œëŠ” í•¨ìˆ˜ ì •ì˜
def select_diag(sigma, k):
    result = np.zeros_like(sigma)  # ì˜í–‰ë ¬ ë§Œë“¤ì–´ë‘ê¸°
    result[k, k] = sigma[k, k]  # sigma í–‰ë ¬ì˜ (k,k) ì›ì†Œë§Œ ë‚¨ê¹€
    return result

# Sigmaì˜ (k,k)ë²ˆì§¸ ì›ì†Œë§Œ ë‚¨ê¸°ê³  ë‚˜ë¨¸ì§€ë¥¼ 0ìœ¼ë¡œ ë§Œë“¤ê¸°
sigma_k = select_diag(Sigma, 3)
# ê²°ê³¼ í–‰ë ¬ ì‹œê°í™”
plot_matrix(sigma_k)
````

í–‰ë ¬ Mì´ rê°œì˜ singular valueë¥¼ ê°€ì§„ë‹¤ê³  í•  ë•Œ

$\Sigma = \Sigma_0 + \Sigma_1 + ... \Sigma_r$ì´ë¯€ë¡œ, ìœ„ì˜ SVDì‹ì„ ë‹¤ìŒê³¼ ê°™ì´ ì ì„ ìˆ˜ ìˆìœ¼ë©°:

$M=U\Sigma V^T = U\Sigma_0V^T + U\Sigma_1V^T + ... U\Sigma_rV^T$,

ì´ ì‹ì„ ì½”ë“œë¡œëŠ” ë‹¤ìŒê³¼ ê°™ì´ í™•ì¸í•´ë³¼ ìˆ˜ ìˆìŠµë‹ˆë‹¤.

````python
# ì›ë˜ í–‰ë ¬ì˜ rank
r = np.linalg.matrix_rank(M)

# ê²°ê³¼ í–‰ë ¬ì„ ë¯¸ë¦¬ initialize
result = np.zeros_like(M)
for k in range(r):
    # Sigma_kë¥¼ ê³„ì‚° í›„ ê²°ê³¼í–‰ë ¬ì— ë”í•´ì¤Œ.
    sigma_k = select_diag(Sigma, k)
    result += U @ sigma_k @ V.T

# ê²°ê³¼ í–‰ë ¬ê³¼ ì›ë˜ í–‰ë ¬ì˜ ì°¨ì´ë¥¼ ê³„ì‚°
print("Maximum diff: ", np.abs(M - result).max())
# ê²°ê³¼ í–‰ë ¬ ì‹œê°í™”
plot_matrix(result)
````

ê·¸ëŸ°ë°, $\Sigma_k$ëŠ” ê²°êµ­ (k, k) ì›ì†Œë§Œ 0ì´ ì•„ë‹Œ í–‰ë ¬ì´ë¯€ë¡œ, ë‹¤ìŒê³¼ ê°™ì´ Uì™€ Vì˜ kë²ˆì§¸ ì—´ë²¡í„°
í•˜ë‚˜ì”©ë§Œ ê°€ì§€ê³ ë„ ê³„ì‚°í•˜ë”ë¼ë„ ì™„ì „íˆ ë™ì¼í•œ ê²°ê³¼ë¥¼ ì–»ì„ ìˆ˜ ìˆìŠµë‹ˆë‹¤.

````python
# kë²ˆì§¸ ì—´ë²¡í„°ë¥¼ ê°€ì ¸ì˜¤ëŠ” í•¨ìˆ˜ë¥¼ ì •ì˜
def col_vec(matrix, k):
    return matrix[:, [k]]

# ì›ë˜ í–‰ë ¬ì˜ rank
r = np.linalg.matrix_rank(M)

# ê²°ê³¼ í–‰ë ¬ì„ ë¯¸ë¦¬ initialize
result = np.zeros_like(M)
for k in range(r):
    # kë²ˆì§¸ singular value, ìŠ¤ì¹¼ë¼ê°’
    sig_k = Sigma[k, k]
    # U, Vì—ì„œ í•œê°œì˜ column vectorë§Œì„ ê°€ì ¸ì™€ ì‚¬ìš©í•´ë„ ìœ„ ì…€ê³¼ ë™ì¼í•œ ê²°ê³¼
    result += sig_k * col_vec(U, k) @ col_vec(V, k).T

# ê²°ê³¼ í–‰ë ¬ê³¼ ì›ë˜ í–‰ë ¬ì˜ ì°¨ì´ë¥¼ ê³„ì‚°
print("Maximum diff: ", np.abs(M - result).max())
# ê²°ê³¼ í–‰ë ¬ ì‹œê°í™”
plot_matrix(result)
````

ì—¬ê¸°ì—ì„œ ê° ë‹¨ê³„ì— ë”í•´ì§„ í–‰ë ¬ë“¤ì²˜ëŸ¼, ë‘ ë²¡í„°ì˜ ê³±ìœ¼ë¡œ í‘œí˜„ë˜ëŠ” í–‰ë ¬ì€ ì„ í˜•ëŒ€ìˆ˜í•™ì—ì„œ rank-1
matrixë¼ê³  ë¶€ë¥´ëŠ” ê°œë…ìœ¼ë¡œ,
í¬ê¸°ëŠ” í–‰ë ¬ì´ì§€ë§Œ ê°€ì§€ê³ ìˆëŠ” ì •ë³´ëŸ‰ì€ ê³±í•´ì§€ëŠ” ë²¡í„° ë‘ê°œì™€ ë™ì¼í•˜ë‹¤ëŠ” íŠ¹ì„±ì„ ê°€ì§€ë©°,
í–‰ë ¬ì„ ë§Œë“œëŠ” ê¸°ë³¸ë‹¨ìœ„ì²˜ëŸ¼ ì‚¬ìš©ë˜ëŠ” ê°œë…ì…ë‹ˆë‹¤.

````python
# ì„ì˜ì˜ kë¥¼ ì„ íƒí•©ë‹ˆë‹¤.
k = 2
# ê³±í•´ì§€ëŠ” U, Vì˜ ì—´ë²¡í„°ë¥¼ í‘œì‹œí•©ë‹ˆë‹¤.
print(col_vec(U, k).shape)
print(col_vec(V, k).T.shape)

# ë‘ ë²¡í„°ì˜ í–‰ë ¬ê³±ìœ¼ë¡œ rank-1 matrixë¥¼ ë§Œë“­ë‹ˆë‹¤.
rank1_mat = col_vec(U, k) @ col_vec(V, k).T

print(rank1_mat.shape)

# ì´ë¦„ì²˜ëŸ¼ rankëŠ” 1ë¡œ ë‚˜ì˜µë‹ˆë‹¤.
print("Rank of resulting Rank-1 Matrix: ", np.linalg.matrix_rank(rank1_mat))

# matrixì˜ ê°’ë“¤ì„ ì‚´í´ë³´ë©´, ì²«ë²ˆì§¸ rowì— ìŠ¤ì¹¼ë¼ê³±ì„ í•´ì„œ ë‚˜ë¨¸ì§€ ëª¨ë“ 
# rowë¥¼ ë§Œë“¤ìˆ˜ ìˆê³ (linearly dependence), ì´ëŠ” columnë„ ë§ˆì°¬ê°€ì§€ì…ë‹ˆë‹¤.
plot_matrix(rank1_mat)
````

## 2. Truncated SVDë¥¼ í†µí•œ í–‰ë ¬ì˜ rank ì¶•ì†Œ

#### ğŸ“ ì„¤ëª…: Truncated SVDë¥¼ í†µí•œ í–‰ë ¬ì˜ rank ì¶•ì†Œ

ìœ„ì—ì„œëŠ” rankê°€ rì¸(rê°œì˜ singular valueë¥¼ ê°€ì§€ëŠ”) í–‰ë ¬ Mì´ ìˆì„ ë•Œ,
SVDë¥¼ í™œìš©í•˜ë©´ ì´ë¥¼ rê°œì˜ rank1 matrixë“¤ì˜ í•©ìœ¼ë¡œ í‘œí˜„í•  ìˆ˜ ìˆë‹¤ëŠ” ê²ƒì„ ì•Œì•„ë³´ì•˜ìŠµë‹ˆë‹¤.
Truncated SVDë€ ì´ë¥¼ í™œìš©í•´, rê°œì˜ ì„±ë¶„ë“¤ ì¤‘ í¬ê¸°ê°€ ì‘ì€ ê²ƒë“¤ì„ ë²„ë¦¬ê³  í–‰ë ¬ Mì„ ë‚®ì€ rankì˜ í–‰ë ¬ë¡œ ë‚˜íƒ€ë‚´ëŠ” ì°¨ì›ì¶•ì†Œ ë°©ë²•ë¡ ì´ë¼ê³  ì´í•´í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.

ë¨¼ì €, ìœ„ì— ì‘ì„±í–ˆë˜ ì½”ë“œë¥¼ ì‚´ì§ ìˆ˜ì •í•´, Truncated SVDë¥¼ í†µí•œ ì°¨ì›ì¶•ì†Œë¥¼ ìˆ˜í–‰í•˜ëŠ” í•¨ìˆ˜ë¥¼ ì •ì˜í•´ë³´ê² ìŠµë‹ˆë‹¤.

````python
def reduce_dim(M, n_components=None):
    # ì£¼ì–´ì§„ í–‰ë ¬ Mì„ SVDí•©ë‹ˆë‹¤.
    U, Sigma, V = full_svd(M)

    r = np.linalg.matrix_rank(M)
    if n_components is None:
        # ëª‡ê°œì˜ ì„±ë¶„ì„ ë‚¨ê¸¸ì§€ ì£¼ì–´ì§€ì§€ ì•Šìœ¼ë©´ ì•„ë¬´ê²ƒë„ ë²„ë¦¬ì§€ ì•Šê³  ì „ì²´ ì„±ë¶„ì„ ë‚¨ê¹ë‹ˆë‹¤.
        n_components = r

    # ë‚¨ê¸¸ component ìˆ˜ê°€ ì „ì²´ ë­í¬ë³´ë‹¤ í¬ë©´ ì—ëŸ¬ ë©”ì‹œì§€ ë³´ì—¬ì¤ë‹ˆë‹¤.
    assert n_components <= r, \
        f"ë‚¨ê¸¸ componentì˜ ê°œìˆ˜({n_components})ëŠ” ì „ì²´ ë­í¬{r}ë³´ë‹¤ í´ ìˆ˜ ì—†ìŠµë‹ˆë‹¤."

    # ê²°ê³¼ í–‰ë ¬ì„ ë¯¸ë¦¬ initialize
    result = np.zeros_like(M, dtype=np.float64)
    # ì´ë²ˆì—ëŠ” rê°œê°€ ì•„ë‹ˆë¼, ì²« n_componentsê°œê¹Œì§€ë§Œ rank-1 matrixë“¤ì„ ë”í•´ì¤ë‹ˆë‹¤.
    for k in range(n_components):
        # kë²ˆì§¸ singular value, ìŠ¤ì¹¼ë¼ê°’ê³¼
        sig_k = Sigma[k, k]
        # rank1 í–‰ë ¬ì— ìœ„ sig_k ìŠ¤ì¹¼ë¼ê°’ì„ ê³±í•´ ê²°ê³¼ì— ë”í•¨.
        result += sig_k * col_vec(U, k) @ col_vec(V, k).T
    return result
````

ì´ì œ ë‹¤ìŒ ì…€ì—ì„œ truncated_svdë¥¼ í†µí•´ ì²« kê°œì˜ ì„±ë¶„ë§Œ ë‚¨ê¸´ í–‰ë ¬ê³¼ ì›ë˜ í–‰ë ¬ì„ ë¹„êµí•´ë³¼ ìˆ˜ ìˆìŠµë‹ˆë‹¤.

````python
# ë‚¨ê¸¸ ì„±ë¶„ì˜ ìˆ˜. ì´ ê°’ì„ 0~6ê¹Œì§€ ì§ì ‘ ê°’ì„ ì¡°ì ˆí•´ë³¼ìˆ˜ ìˆìŠµë‹ˆë‹¤.
n_components = 5
size_scale = 0.6 # ë¹„êµí•´ë³´ê¸° í¸í•˜ë„ë¡ figureí¬ê¸°ë¥¼ ì‚´ì§ ì¤„ì…ë‹ˆë‹¤.

# ì›ë˜ í–‰ë ¬ì˜ rankë¥¼ êµ¬í•´ë´…ë‹ˆë‹¤.
print("Original rank: ", np.linalg.matrix_rank(M))

# TruncatedSVDë¥¼ ìˆ˜í–‰í•´ n_componentê°œì˜ ì„±ë¶„ë§Œ ë‚¨ê¹ë‹ˆë‹¤.
result = reduce_dim(M, n_components)

# TruncatedSVD ê²°ê³¼ í–‰ë ¬ì˜ rankì™€ ì›ë³¸ í–‰ë ¬ê³¼ì˜ ê°’ ì°¨ì´ë¥¼ í”„ë¦°íŠ¸í•©ë‹ˆë‹¤.
print("Result rank: ", np.linalg.matrix_rank(result))
print("Maximum diff: ", np.abs(M - result).max())
````

````python
# ì›ë˜ì˜ í–‰ë ¬
plot_matrix(M, size_scale=size_scale)
plt.title('original matrix M')  # êµ¬ë¶„ì„ ìœ„í•´ ì‹œê°í™” ì œëª© ë„£ê¸°
plt.show()                      # ê·¸ë¦¼ ë³´ì—¬ì£¼ê¸°
````

````python
# truncated svd ê²°ê³¼ í–‰ë ¬
plot_matrix(result, size_scale=size_scale)
plt.title('truncated SVD result')  # êµ¬ë¶„ì„ ìœ„í•´ ì‹œê°í™” ì œëª© ë„£ê¸°
plt.show()               # ê·¸ë¦¼ ë³´ì—¬ì£¼ê¸°
````

````python
# ì›ë˜ í–‰ë ¬ê³¼ì˜ ì°¨ì´(absolute diff)
plot_matrix(np.abs(M - result), size_scale=size_scale)
plt.title('absolute diff: |M-trSVD|')  # êµ¬ë¶„ì„ ìœ„í•´ ì‹œê°í™” ì œëª© ë„£ê¸°
plt.show()                  # ê·¸ë¦¼ ë³´ì—¬ì£¼ê¸°
````

#### ğŸ“ ì„¤ëª…: EigenFace

EigenFaceëŠ” 1990ë…„ëŒ€ ì—°êµ¬ë˜ì—ˆë˜ ì–¼êµ´ì¸ì‹(Face Recognition)ì„ ìœ„í•œ ë°©ë²•ë¡ ìœ¼ë¡œ, PCAë¥¼
ì´ìš©í•´ ì‚¬ëŒì˜ ì–¼êµ´ ì´ë¯¸ì§€ë¥¼ ì €ì°¨ì›ì˜ ë²¡í„°ë¡œ í‘œí˜„í•˜ê³ , ì´ë¥¼ ì„œë¡œ ë¹„êµí•´ ê°™ì€ ì‚¬ëŒì¸ì§€ ì•„ë‹Œì§€ íŒë‹¨í•˜ëŠ” ë°©ì‹ìœ¼ë¡œ ì‘ë™í•©ë‹ˆë‹¤. ì—¬ê¸°ì„œëŠ” The Olivetti faces ë°ì´í„°ì…‹ ([ë§í¬](https://scikit-learn.org/stable/modules/generated/sklearn.datasets.fetch_olivetti_faces.html))ì„
ê°€ì ¸ì™€, Eigen Faceì˜ ì°¨ì›ì¶•ì†Œê°€ ì–´ë–»ê²Œ ì‘ë™í•˜ëŠ”ì§€ ì‹¤ìŠµí•´ë´…ë‹ˆë‹¤.

#### ğŸ“š ìë£Œ

*  [Faces dataset decompositions](https://scikit-learn.org/stable/auto_examples/decomposition/plot_faces_decomposition.html#sphx-glr-auto-examples-decomposition-plot-faces-decomposition-py) : scikit
learnì—ì„œ ì œê³µí•˜ëŠ” ê³µì‹ íŠœí† ë¦¬ì–¼ ë¬¸ì„œë¡œ, ì´ ì‹¤ìŠµì½”ë“œì—ì„œ Olivetti faces ë°ì´í„°ì…‹ì˜ ì²˜ë¦¬ì™€ ì‹œê°í™”
ë¶€ë¶„ì€ í•´ë‹¹ ë¬¸ì„œì˜ ì½”ë“œë¥¼ ì°¸ê³ í•˜ì—¬ ì‘ì„±í–ˆìŠµë‹ˆë‹¤. ì´ ì‹¤ìŠµì½”ë“œê°€ PCAë¼ëŠ” í•œê°€ì§€ ë°©ë²•ë¡ ì˜ ì›ë¦¬ë¥¼
ì„¤ëª…í•˜ëŠ”ë°ì— ì§‘ì¤‘í–ˆë‹¤ë©´, ìœ„ ë¬¸ì„œì—ì„œëŠ” ê°™ì€ ë°ì´í„°ì— ë‹¤ì–‘í•œ ë°©ì‹ì˜ ì°¨ì›ì¶•ì†Œë¥¼ ì ìš©í•´ ì–´ë–¤ ì°¨ì´ê°€ ë‚˜íƒ€ë‚˜ëŠ”ì§€ ì˜ˆì‹œë¥¼ í™•ì¸í•´ë³¼ ìˆ˜ ìˆìŠµë‹ˆë‹¤.

ë°ì´í„°ì˜ ê° ê°’ë“¤ì€ 0ì—ì„œ 1 ì‚¬ì´ì˜ ê°’ì„ ê°€ì§‘ë‹ˆë‹¤.



ì´ ë°ì´í„°ë¥¼ ìš°ë¦¬ê°€ ì•Œì•„ë³¼ ìˆ˜ ìˆëŠ” ì´ë¯¸ì§€ì˜ í˜•íƒœë¡œ í™•ì¸í•˜ë ¤ë©´ ë³„ë„ì˜ ë³€í™˜ ê³¼ì •ì„ ê±°ì³ì•¼ í•˜ëŠ”ë°,
í•œë²ˆ 400ê°œì¤‘ í•˜ë‚˜ì˜ ì´ë¯¸ì§€ë¥¼ ì„ íƒí•´ì„œ, ì´ë¯¸ì§€ë¡œ ë³€í™˜í•´ë³´ê² ìŠµë‹ˆë‹¤.

````python
# ê°€ì ¸ì˜¬ ìƒ˜í”Œì˜ ë²ˆí˜¸ë¥¼ ëœë¤ìœ¼ë¡œ ê³ ë¦„
index = np.random.choice(len(faces))
print(f"{index}-th row of matrix")

# ì›ë³¸ ì´ë¯¸ì§€ì˜ í¬ê¸°
img_h, img_w = (64, 64)

# ë°ì´í„°(faces)ì—ì„œ ìƒ˜í”Œ(row)ì„ ì„ íƒí•´ ê°€ì ¸ì˜¤ê¸°
face_vector = faces[index]

# ì´ë¯¸ì§€ë¥¼ ì›ë˜ì˜ í¬ê¸°ë¡œ ë³€í™˜í•œ í›„ display
face_image = face_vector.reshape(img_h, img_w)
plt.imshow(face_image, cmap="gray")
````

ì§€ê¸ˆì€ matplotlibì˜ `imshow`í•¨ìˆ˜ê°€ **0\~1 ë²”ìœ„ì˜ float ìˆ«ì**ë„ ì´ë¯¸ì§€ë¡œ ì‰½ê²Œ ë³€í™˜í•´ë³¼ ìˆ˜ ìˆëŠ” ê¸°ëŠ¥ì„ ì§€ì›í•˜ê¸° ë•Œë¬¸ì—, ê°’ì˜ ë²”ìœ„ë¥¼ **0\~255 ë²”ìœ„ì˜ í”½ì…€ê°’(int)**ìœ¼ë¡œ ë³€í™˜í•˜ëŠ” ì‘ì—…ì€ í•˜ì§€ ì•Šì•„ë„ ê´œì°®ìŠµë‹ˆë‹¤.
í•˜ì§€ë§Œ PCAë¶„ì„ì„ ìœ„í•´ì„œëŠ” ë‹¤ìŒ ê° ë°ì´í„°ì˜ feature-wise, sample-wiseë¡œ í‰ê· ì´ ëª¨ë‘ 0ì´
ë˜ì–´ì•¼í•˜ë¯€ë¡œ ì´ì— ë§ì¶° ë‹¤ìŒê³¼ ê°™ì´ ì „ì²˜ë¦¬ë¥¼ ìˆ˜í–‰í•´ì¤€ ì´í›„ì—ëŠ” ë³„ë„ì˜ scaling ê³¼ì •ì„ ê±°ì³ì•¼ ì œëŒ€ë¡œ
ì´ë¯¸ì§€ë¥¼ í™•ì¸í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.

````python
# ì „ì²´ ìƒ˜í”Œë‹¨ìœ„ì˜ í‰ê· ì„ êµ¬í•˜ê³ , ì´ë¥¼ ì›ë³¸ ë°ì´í„°ì—ì„œ ë¹¼ì„œ í‰ê· ì„ 0ìœ¼ë¡œ ë§ì¶°ì¤Œ
samplewise_mean = faces.mean(axis=0) # (4096, )
faces_centered = faces - samplewise_mean

# ê° ì´ë¯¸ì§€ë§ˆë‹¤ ëª¨ë“  í”½ì…€ê°’ì˜ í‰ê· ì„ êµ¬í•˜ê³ , ì´ë¥¼ ì›ë³¸ ì´ë¯¸ì§€ì—ì„œ ë¹¼ëŠ” ë°©ì‹ìœ¼ë¡œ í‰ê· ì„ 0ìœ¼ë¡œ ë§ì¶°ì¤Œ
pixelwise_mean = faces_centered.mean(axis=1).reshape(n_samples, -1) # (400, )
faces_centered -= pixelwise_mean
````

````python
# ì‹œê°í™”ë¥¼ ìœ„í•œ í•¨ìˆ˜ ì •ì˜í•˜ê¸°
def plot_faces(title, images, n_cols=3, n_rows=2, shuffle=False, cmap="gray", size_scale=2.0, random_seed=0, image_shape=(64, 64)):
    # plotí•  ì´ë¯¸ì§€(ë²¡í„°)ë“¤ì„ ëœë¤ìœ¼ë¡œ ì„ íƒ
    if shuffle:
        np.random.seed(random_seed)
        indices = np.random.choice(len(images), n_cols * n_rows)
    else:
        indices = np.arange(n_cols * n_rows)

    # figureê´€ë ¨ ì„¤ì •
    fig, axs = plt.subplots(
        nrows=n_rows,
        ncols=n_cols,
        figsize=(n_cols * size_scale, n_rows * size_scale),
        facecolor="white",
        constrained_layout=True,
    )
    fig.set_constrained_layout_pads(w_pad=0.01, h_pad=0.02, hspace=0, wspace=0)
    fig.set_edgecolor("black")
    fig.suptitle(title, size=16)

    # ê° ìë¦¬ì— ë“¤ì–´ê°€ëŠ” ì–¼êµ´ ì´ë¯¸ì§€ë¥¼ plot
    for ax, idx in zip(axs.flat, indices):
        face_vec = images[idx]
        vmax = max(face_vec.max(), - face_vec.min())
        im = ax.imshow(
            face_vec.reshape(image_shape),
            cmap=cmap,
            interpolation="nearest",
            vmin=-vmax,
            vmax=vmax,
        )
        ax.axis("off")
    fig.colorbar(im, ax=axs, orientation="horizontal", shrink=0.99, aspect=40, pad=0.01)
    plt.show()
````

````python
# ì´ë¯¸ì§€ 6ê°œ ì‹œê°í™”í•˜ê¸°
plot_faces("Faces from dataset", faces_centered, shuffle=True, random_seed=1234)
````

````python
# ì¤„ì¼ ì°¨ì›ì˜ ìˆ˜ ì§€ì •í•˜ê¸°
n_components = 20

# PCA ìˆ˜í–‰í•˜ê¸°
pca_estimator = PCA(n_components=n_components, svd_solver="full", whiten=True)
pca_estimator.fit(faces_centered)

# PCA ê²°ê³¼ (Eigenface) ì‹œê°í™”
plot_faces("Components of PCA", pca_estimator.components_, n_rows=2, n_cols=4)
````

ìœ„ì—ì„œ ë°©ê¸ˆ ìˆ˜í–‰í•œ ì‘ì—…ì€ 400ê°œì˜ ì–¼êµ´ ì´ë¯¸ì§€ë“¤ì„ ì„œë¡œ êµ¬ë¶„í•˜ëŠ”ë° ê°€ì¥ ìœ ì˜ë¯¸í•œ ì²« 8ê°œì˜
componentsë“¤ì„ ë‹¤ì‹œ ì´ë¯¸ì§€ì˜ í˜•íƒœë¡œ ì‹œê°í™”í•œê²ƒì…ë‹ˆë‹¤.

ì´ë¥¼ ì´í•´í•˜ê¸° ìœ„í•´ 400ê°œì˜ ì´ë¯¸ì§€ë“¤ ì¤‘ í•˜ë‚˜ë¥¼ ê³¨ë¼ ì°¨ì›ì¶•ì†Œëœ ë²¡í„°ì™€ componentsë¡œ ì›ë³¸ ì´ë¯¸ì§€ë¥¼ ë³µì›í•´ë³´ë„ë¡ í•˜ê² ìŠµë‹ˆë‹¤.

````python
# ì›ë³¸ ì´ë¯¸ì§€ì¤‘ì— ì„ì˜ë¡œ í•˜ë‚˜ë¥¼ ê³ ë¦…ë‹ˆë‹¤.
index = 123
indices = np.random.choice(n_samples, 6)
# ì›ë³¸ ì´ë¯¸ì§€ë¥¼ ë³´ì—¬ì¤ë‹ˆë‹¤.
plt.title(f"Original Face Image at index: {index}")
plt.imshow(faces[index].reshape(64, 64), cmap="gray")
````

ì´ ì´ë¯¸ì§€ëŠ” ì›ë˜ì˜ 4096(64*64)ì°¨ì›ì—ì„œ, PCA estimatorë¥¼ í†µí•´ ì°¨ì›ì¶•ì†Œë˜ì–´ ë‹¤ìŒê³¼ ê°™ì€ `n_components` ì°¨ì› ë²¡í„°ë¡œ í‘œí˜„ë©ë‹ˆë‹¤.

````python
# ì°¨ì›ì¶•ì†Œëœ ë²¡í„° ê³„ì‚°í•˜ê¸°
reduced_vec = pca_estimator.transform(faces_centered[index].reshape(1, -1))
print(reduced_vec)
print('ì°¨ì› ì¶•ì†Œëœ ë²¡í„°ì˜ í¬ê¸°:', reduced_vec.shape)
````

fitting ëœ PCAì˜ componentë“¤ì„, ìœ„ reduced_vecì˜ ì›ì†Œë¥¼ ê³„ìˆ˜ë¡œ í•˜ëŠ” ì„ í˜•ê²°í•©í•˜ë©´,
ë‹¤ìŒê³¼ ê°™ì´ ì›ë³¸ ì´ë¯¸ì§€ë¥¼ ë³µì›í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.

````python
# ê²°ê³¼ í–‰ë ¬ ë¯¸ë¦¬ initialize
canvas = np.zeros([64, 64], dtype=np.float64)
for value, comp in zip(reduced_vec[0], pca_estimator.components_):
    # ê° component ë²¡í„°ë¥¼ ì´ë¯¸ì§€ í¬ê¸°ë¡œ resizeí•œ ë’¤, ì´ë¥¼ ì°¨ì›ì¶•ì†Œëœ ë²¡í„°ì˜ ê° ê°’ê³¼ ì„ í˜•ê²°í•©
    canvas += comp.reshape(64, 64) * value

vmax = max(canvas.max(), - canvas.min())

plt.imshow(canvas, cmap="gray", vmax=vmax, vmin=-vmax)
````

ì§€ê¸ˆì€ ì¶•ì†Œëœ ì°¨ì›ì˜ ìˆ˜, `n_components`ê°€ ë„ˆë¬´ ì‘ì•„ ë³µì›ëœ ì´ë¯¸ì§€ê°€ ì›ë³¸ê³¼ ê°™ì€ ì´ë¯¸ì§€ë¼ëŠ” ê²ƒì€
ê²¨ìš° ì•Œì•„ë³¼ ìˆ˜ ìˆì„ ì •ë„ì§€ë§Œ, ì›ë³¸ ì´ë¯¸ì§€ì˜ ë””í…Œì¼í•œ ë¶€ë¶„ë“¤ì€ ìƒë‹¹íˆ ì†Œì‹¤ëœ ìƒíƒœì…ë‹ˆë‹¤.
`n_components` ê°’ì— ë”°ë¼ ì°¨ì›ì¶•ì†Œëœ ì´ë¯¸ì§€ê°€ ì–´ë–¤ í˜•íƒœë¡œ ë‚˜íƒ€ë‚˜ëŠ”ì§€ ì•„ë˜ì—ì„œ ë¹„êµí•´ë´…ë‹ˆë‹¤.

````python
# ì›ë³¸ ì´ë¯¸ì§€ì™€ ì°¨ì›ì¶•ì†Œëœ ì´ë¯¸ì§€ë“¤ ë¹„êµí•˜ê¸°
def compare_reduced_faces(title, images, index=123, n_components_list=[5, 20, 100], n_cols=4, n_rows=1, shuffle=False, cmap="gray", size_scale=2.5, random_seed=0, image_shape=(64, 64)):
    # ê·¸ë¦¼ ê´€ë ¨ ì„¤ì •
    fig, axs = plt.subplots(
        nrows=n_rows, ncols=n_cols,
        figsize=(n_cols * size_scale, n_rows * size_scale),
        facecolor="white",
        constrained_layout=True,
    )
    fig.set_constrained_layout_pads(w_pad=0.01, h_pad=0.02, hspace=0, wspace=0)
    fig.set_edgecolor("black")

    # ë³´ì—¬ì¤„ ì´ë¯¸ì§€ ì„ ì •
    face_vec = faces[index]

    # ì²« ì´ë¯¸ì§€ë¡œ ì›ë³¸ ì´ë¯¸ì§€ë¥¼ ë³´ì—¬ì¤ë‹ˆë‹¤.
    axs[0].set_title("Original Face Image", y=-0.2)
    axs[0].imshow(face_vec.reshape(image_shape), cmap="gray")
    axs[0].axis("off")

    # ë‹¤ìŒ ì´ë¯¸ì§€ë¶€í„°ëŠ” PCAë¥¼ ì´ìš©í•´ ì°¨ì›ì¶•ì†Œëœ ì´ë¯¸ì§€ë¥¼ ë³´ì—¬ì¤ë‹ˆë‹¤.
    # ê° ì°¨ì›ë§ˆë‹¤ ë³´ì—¬ì£¼ë¯€ë¡œ ì¤„ì¼ ì°¨ì›ì˜ ìˆ˜ ë¦¬ìŠ¤íŠ¸ ì¤‘ í•˜ë‚˜ì”© ì§€ì •í•˜ì—¬ PCAë¥¼ ìˆ˜í–‰í•©ë‹ˆë‹¤.
    for img_index, n_components in enumerate(n_components_list):

        # PCA ìˆ˜í–‰í•˜ê¸°
        pca_estimator = PCA(n_components=n_components, svd_solver="full", whiten=True)
        pca_estimator.fit(images)

        # ì°¨ì›ì¶•ì†Œëœ ë²¡í„° ê³„ì‚°í•˜ê¸°
        reduced_vec = pca_estimator.transform(face_vec.reshape(1, -1))
        # ê²°ê³¼ í–‰ë ¬ ë¯¸ë¦¬ initialize
        canvas = np.zeros([64, 64], dtype=np.float64)
        for value, comp in zip(reduced_vec[0], pca_estimator.components_):
            # ê° component ë²¡í„°ë¥¼ ì´ë¯¸ì§€ í¬ê¸°ë¡œ resizeí•œ ë’¤, ì´ë¥¼ ì°¨ì›ì¶•ì†Œëœ ë²¡í„°ì˜ ê° ê°’ê³¼ ì„ í˜•ê²°í•©
            canvas += comp.reshape(64, 64) * value

        # PCA ê²°ê³¼ (Eigenface) ì‹œê°í™”
        vmax = max(canvas.max(), - canvas.min())
        im = axs[img_index+1].imshow(
            canvas.reshape(image_shape),
            cmap=cmap,
            interpolation="nearest",
            vmin=-vmax,
            vmax=vmax,
        )
        axs[img_index+1].axis("off")
        axs[img_index+1].set_title(f'Dimension={n_components}', y=-0.2)

    # ìµœì¢… ì´ë¯¸ì§€ ë³´ì—¬ì£¼ê¸°
    plt.suptitle(title + f': images at index {index}', fontsize=20)
    plt.show()
````

````python
compare_reduced_faces('Comparisons of different dimensions', faces_centered, n_components_list=[5, 20, 100])
````

`n_components`ê°€ 5ì¼ ë•Œ, ì¦‰ 5ì°¨ì›ìœ¼ë¡œ ì¶•ì†Œì‹œí‚¤ë©´ ì‚¬ëŒì´ë¼ëŠ” ì´ë¯¸ì§€ëŠ” ì¸ì‹í•  ìˆ˜ ìˆìœ¼ë‚˜ ì›ë³¸ ì´ë¯¸ì§€ì— ë‚˜íƒ€ë‚œ ì‚¬ëŒê³¼ëŠ” ë‹¤ì†Œ ì°¨ì´ê°€ ìˆê³  ì•ˆê²½ë„ ì¸ì‹í•  ìˆ˜ ì—†ìŠµë‹ˆë‹¤.
`n_components`ê°€ 100ì¼ ë•ŒëŠ” ì›ë³¸ ì´ë¯¸ì§€ì˜ ì—¬ëŸ¬ ë””í…Œì¼í•œ ì •ë³´ë“¤ê¹Œì§€ í•¨ê»˜ ë³µì›ë˜ì—ˆìŠµë‹ˆë‹¤.
ì´ì²˜ëŸ¼ `n_components` ê°’ì„ ëŠ˜ë¦¬ë©´ ì¢€ë” ì„¸ì„¸í•œ ì •ë³´ë“¤ê¹Œì§€ ìœ ì§€í•˜ë©´ì„œ ì°¨ì›ì¶•ì†Œë¥¼ ì§„í–‰í•  ìˆ˜ ìˆì§€ë§Œ, ëŒ€ì‹  ê·¸ë§Œí¼ ì°¨ì› ìˆ˜ê°€
ëœ ì¤„ì–´ë“œëŠ” ê²ƒì´ë¯€ë¡œ ì–´ëŠì •ë„ ì ë‹¹í•œ ì„ ì—ì„œ ê²°ì •í•´ì•¼ í•©ë‹ˆë‹¤.

ì–´ëŠì •ë„ì˜ `n_components`ì˜ ê°’ì´ ì ì ˆí•œì§€ëŠ” ë°ì´í„°ê°€ ê°€ì§€ê³ ìˆëŠ” ì •ë³´ëŸ‰, ì¦‰
ì–¼ë§ˆë‚˜ ì˜ˆì¸¡ ê°€ëŠ¥í•œ ì •ë³´ì¸ê°€ì— ë”°ë¼ ë‹¬ë¼ì§‘ë‹ˆë‹¤. ì˜ˆì‹œë¥¼ ìœ„í•´ ë§¤ìš° ë†’ì€ "ì •ë³´ëŸ‰"ì„ ê°€ì§„ ëœë¤ í–‰ë ¬ì„ ë§Œë“¤ì–´
PCAë¡œ í”¼íŒ…í–ˆì„ ë•Œ ìœ„ì˜ ê²°ê³¼ì™€ ì–´ë–¤ ì°¨ì´ê°€ ìˆëŠ”ì§€ ì‚´í´ë³´ê² ìŠµë‹ˆë‹¤.

ë°©ê¸ˆì˜ PCAì— ì‚¬ìš©ëœ singular values, ì¦‰ $\Sigma$ì˜ ëŒ€ê° ì›ì†Œë“¤ì€ ë‹¤ìŒê³¼ ê°™ìŠµë‹ˆë‹¤. ì´ ê°’ë“¤ì€ ê°
ì„±ë¶„, ì¦‰ componentê°€ ì´ë¯¸ì§€ë“¤ì„ ì„œë¡œ êµ¬ë¶„í•˜ëŠ”ë° ì–¼ë§ˆë‚˜ ì¤‘ìš”í•œì§€ë¥¼ ë‚˜íƒ€ë‚´ëŠ”ë°, ë’¤ë¡œê°ˆìˆ˜ë¡ ë¹ ë¥´ê²Œ ê°’ì´
ì‘ì•„ì§€ëŠ” ê²ƒ, ì¦‰ í•´ë‹¹ ì„±ë¶„ì´ ì´ë¯¸ì§€ì˜ êµ¬ë¶„ì— ëœ ì¤‘ìš”í•˜ë‹¤ëŠ” ê²ƒì„ í™•ì¸í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.

````python
# singular value í™•ì¸í•˜ê¸°
pca_estimator.singular_values_
````

ë¹„êµë¥¼ ìœ„í•´, ìœ„ì˜ ì–¼êµ´ ë°ì´í„°ì™€ í¬ê¸°, ëª¨ì–‘ì´ ê°™ì§€ë§Œ ê°’ ìì²´ëŠ” ê·¸ëƒ¥ ëœë¤ì •ê·œë¶„í¬ì—ì„œ ìƒ˜í”Œë§í•œ ê°€ì§œ ë°ì´í„°
í–‰ë ¬ì„ ë§Œë“¤ì–´ PCA fittingì„ ìˆ˜í–‰í•´ë³´ê² ìŠµë‹ˆë‹¤.

````python
# ê¸°ì¡´ facesë°ì´í„°ì…‹ê³¼ ê°™ì€ í¬ê¸°ì˜ ëœë¤ ì •ê·œë¶„í¬ ë°ì´í„°ë¥¼ ìƒì„±
random_noises = np.random.randn(*faces_centered.shape)

# ì´ ëœë¤ í¬ì¸íŠ¸ì— ëŒ€í•´ PCA ìˆ˜í–‰í•˜ê¸°
pca_estimator = PCA(n_components=n_components, svd_solver="full", whiten=True)
pca_estimator.fit(random_noises)
````

í™•ì—°íˆ ì¤„ì–´ë“œëŠ” ì–¼êµ´ë°ì´í„°ì˜ singular valuesì— ë¹„í•´ ëœë¤í–‰ë ¬ì˜ singular valuesëŠ” ì¤„ì–´ë“¤ê¸°ëŠ”
í•˜ì§€ë§Œ ê°ì†Œí­ì´ í›¨ì”¬ ì•½í•œê²ƒì„ ë³¼ìˆ˜ ìˆìŠµë‹ˆë‹¤.

````python
# singular value í™•ì¸í•˜ê¸°
pca_estimator.singular_values_
````

`n_components`ë¥¼ ìµœëŒ€ë¡œ ì„¤ì •í•´ ë” ë§ì€ ê°’ë“¤ì„ í™•ì¸í•´ë´ë„ ë™ì¼í•œ ê²°ê³¼ê°€ ë‚˜ì˜µë‹ˆë‹¤. ì¦‰, ëœë¤í•˜ê²Œ ìƒì„±ëœ ë°ì´í„° í–‰ë ¬ì˜ signular valueì˜ ê°ì†Œí­ë³´ë‹¤ ì–¼êµ´ ë°ì´í„°ì˜ ê°ì†Œí­ì´ í›¨ì”¬ í½ë‹ˆë‹¤.

````python
# ì–¼êµ´ ë°ì´í„°ì— ëŒ€í•´ ìµœëŒ€ n_componentsë¡œ PCA ìˆ˜í–‰í•˜ê¸°
pca_faces = PCA(n_components=400, svd_solver="full", whiten=True)
pca_faces.fit(faces_centered)
sv_faces = pca_faces.singular_values_

# ëœë¤ ë°ì´í„°ì— ëŒ€í•´ ìµœëŒ€ n_componentsë¡œ PCA ìˆ˜í–‰í•˜ê¸°
pca_random = PCA(n_components=400, svd_solver="full", whiten=True)
pca_random.fit(random_noises)
sv_random = pca_random.singular_values_


# ê°ê°ì˜ singular valueë“¤ì„ plotí•´ ë¹„êµ.
plt.subplot(121)
plt.title("Singular Values(Random)")
plt.xlim(0, 399)
plt.ylim(0, 100)
plt.plot(sv_random)

plt.subplot(122)
plt.title("Singular Values(Faces)")
plt.xlim(0, 399)
plt.ylim(0, 100)
plt.plot(sv_faces)
````

ì´ëŠ” PCAê°€ ìˆ˜í–‰í•˜ëŠ” ì‘ì—…ì´ ê²°êµ­ ìƒ˜í”Œ, í”¼ì³ ê°„ì˜ ìƒí˜¸ ì—°ê´€ì„±, ì¦‰, ê°™ì€ ì´ë¯¸ì§€ì˜ ë‹¤ë¥¸ ë¶€ë¶„ì´ë‚˜ ë°ì´í„°ì…‹ ë‚´ì˜ ë‹¤ë¥¸ ì´ë¯¸ì§€ë¥¼ ê°€ì§€ê³  ì–¼ë§ˆë‚˜ ì‰½ê²Œ ì´ë¯¸ì§€ì˜ íŠ¹ì • ë¶€ë¶„ì„ ì¶”ì¸¡í•  ìˆ˜ ìˆëŠëƒì— ë”°ë¼ ë‹¬ë¼ì§€ê²Œ ëœë‹¤ëŠ” ê²ƒì„ ì˜ë¯¸í•©ë‹ˆë‹¤.

ì‚¬ëŒì˜ ì–¼êµ´ ì´ë¯¸ì§€ëŠ” ì„œë¡œ ë§ì´ ë‹¤ë¥´ì§€ë§Œ ìƒë‹¹íˆ ë§ì€ ìœ ì‚¬ì„±ì„ ê°–ê³  ìˆê¸°ì— ì²« ëª‡ ê°œì˜ componentsì˜ ì˜í–¥ì„ í¬ê²Œ ë°›ëŠ” ê²ƒì´ê³  ê·¸ë˜ì„œ ì°¨ì›ì¶•ì†Œë¥¼ ë§ì´ í•´ì„œ ì„±ë¶„ì˜ ìˆ˜ë¥¼ ë§ì´ ì¤„ì´ë”ë¼ë„ ì›ë³¸ ì´ë¯¸ì§€ê°€ ì•Œì•„ë³¼ ìˆ˜ ìˆê²Œ ë³µì›ë©ë‹ˆë‹¤.

ì™„ì „ ëœë¤ìœ¼ë¡œ ìƒì„±í•œ í–‰ë ¬ì˜ ê²½ìš°ëŠ” ë°ì´í„°ì…‹ì˜ í•œ ì˜ì—­ìœ¼ë¡œë¶€í„° ë‹¤ë¥¸ ì˜ì—­ì„ ë³µì›í•˜ëŠ” ê²ƒì´ ê±°ì˜ ë¶ˆê°€ëŠ¥ì— ê°€ê¹ê¸° ë•Œë¬¸ì—, ì°¨ì›ì„ ì¤„ì˜€ì„ ë•Œ ì†ì‹¤ë˜ëŠ” ì •ë³´ì˜ ë¹„ìœ¨ì´ ë§¤ìš° í¬ê³ , ì°¨ì›ì¶•ì†Œê°€ ê±°ì˜ ë™ì‘í•˜ì§€ ì•ŠëŠ”ë‹¤ê³  ë³´ì•„ë„ ë¬´ë°©í•©ë‹ˆë‹¤.

ì‹¤ì œ í˜„ì‹¤ì˜ ë°ì´í„°ì— PCAë¥¼ ìˆ˜í–‰í• ë•ŒëŠ” ì´ëŸ° ë°ì´í„°ì˜ íŒ¨í„´ê³¼ ì˜ˆì¸¡ ê°€ëŠ¥ì„±ì„ ë³´ê³  ëª‡ ê°œ ì •ë„ì˜ componentsë¥¼ ì‚¬ìš©í• ì§€ ê²°ì •í•˜ê²Œ ë©ë‹ˆë‹¤.

## 4. t-SNE(t-distributed Stochastic Neighbor Embedding)ë¥¼ ì´ìš©í•œ ë°ì´í„°ì…‹ ì‹œê°í™”

```
ğŸ’¡ ëª©ì°¨ ê°œìš” : t-SNE(t-distributed Stochastic Neighbor Embedding)ë¥¼ ì‚¬ìš©í•´ ê³ ì°¨ì› ë°ì´í„°ë¥¼ 2ì°¨ì›ì— ì‹œê°í™”í•  ìˆ˜ ìˆë‹¤.
```

- 4-1. ë°ì´í„° ë¶ˆëŸ¬ì˜¤ê¸°
- 4-2. ë°ì´í„° ì‹œê°í™”



#### ğŸ“ ì„¤ëª…: t-SNE(t-distributed Stochastic Neighbor Embedding)

ê³ ì°¨ì›ì˜ ë°ì´í„°ë¥¼ ìœ ì˜ë¯¸í•˜ê²Œ ì¶•ì†Œì‹œí‚¨ ì°¨ì› ì¶•ì†Œë¥¼ ì´ìš©í•˜ì—¬ ìš°ë¦¬ ëˆˆì— ì´í•´í•˜ê¸° ì‰¬ìš´ í˜•íƒœë¡œ ì „ë‹¬í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤. ì‹œê°í™”ë¥¼ ìœ„í•´ t-SNE (t-distributed Stochastic Neighbor Embedding)ì„ ì´ìš©í•©ë‹ˆë‹¤.

ì˜ˆë¥¼ ë“¤ì–´, 0-9 ìˆ«ì ì´ë¯¸ì§€ ë°ì´í„°ëŠ” 1x64x64 ì°¨ì›ì„ ê°€ì§€ê³  ìˆìŠµë‹ˆë‹¤. ì´ë¯¸ì§€ë¥¼ ì‹œê°í™”í•˜ëŠ” ê²ƒì€ í¬ê²Œ ì–´ë µì§€ ì•Šì§€ë§Œ, ì´ ë°ì´í„°ì˜ í–‰ë ¬ ìì²´ë¥¼ ì‹œê°í™”í•˜ëŠ” ê²ƒì€ êµ‰ì¥íˆ ì–´ë µìŠµë‹ˆë‹¤. ìš°ë¦¬ëŠ” ì´ ìˆ«ì ë°ì´í„°ë¥¼ êµ¬ë¶„í•˜ê¸° ì‰¬ìš´ í˜•íƒœë¡œ ì‹œê°í™”í•˜ê³ ì í•  ë•Œ t-SNEì™€ ê°™ì€ ì‹œê°í™” ë°©ë²•ì„ ì´ìš©í•©ë‹ˆë‹¤.

t-SNEëŠ” ë°ì´í„° ê°„ ìœ ì‚¬ì„±ì„ í™•ë¥ ë¡œ ë³€í™˜í•©ë‹ˆë‹¤. ìœ ì‚¬ì„±ì„ ì´ìš©í•˜ë©´ **ê³ ì°¨ì› ê³µê°„ì—ì„œ ë¹„ìŠ·í•œ ë°ì´í„°ëŠ” ì €ì°¨ì› ê³µê°„ì—ì„œ ê°€ê¹ê³ , ë‹¤ë¥¸ ë°ì´í„°ëŠ” ì„œë¡œ ë©€ë¦¬ ë–¨ì–´ì ¸ìˆë„ë¡ í•™ìŠµ**í•©ë‹ˆë‹¤. ë”°ë¼ì„œ ì„œë¡œ ìœ ì‚¬í•˜ê³  ë‹¤ë¥¸ ê° ë°ì´í„°ë¥¼ 'êµ¬ë¶„'í•˜ê³ ì í•  ë•Œ ë§ì´ ì‚¬ìš©ë  ìˆ˜ ìˆìŠµë‹ˆë‹¤. ìœ„ì˜ ìˆ«ì ì´ë¯¸ì§€ ë°ì´í„° ì˜ˆì‹œì—ì„œì²˜ëŸ¼ ë°ì´í„°ê°€ ì‚°í¬ëœ ë¡œì»¬ êµ¬ì¡°(ì¢ì€ ì˜ì—­)ë¥¼ ê¸°ë°˜ìœ¼ë¡œ ìƒ˜í”Œì„ ê·¸ë£¹í™”í•˜ì—¬ í•œ ë²ˆì— ì—¬ëŸ¬ ë§¤ë‹ˆí´ë“œë¡œ êµ¬ì„±ëœ ë°ì´í„°ë“¤ì„ ì‹œê°ì ìœ¼ë¡œ ë¶„ë¦¬í•˜ëŠ” ë° ë„ì›€ì´ ë  ìˆ˜ ìˆìŠµë‹ˆë‹¤.

ë°ì´í„°ë¥¼ ìƒˆë¡œìš´ ë‹¤ë¥¸ ë‚®ì€ ì°¨ì›ìœ¼ë¡œ ë³€í™˜í•˜ë©´ ì„ë² ë”©(embedding)í–ˆë‹¤ê³  í•˜ëŠ”ë°, ì´ ì„ë² ë”©ëœ(embedded) ê³µê°„ì—ì„œ ë°ì´í„° ê°„ ìœ ì‚¬ì„±ì€ Student t-ë¶„í¬ë¡œ í‘œí˜„ë©ë‹ˆë‹¤. ê·¸ë¦¬ê³  ì›ë˜ ê³µê°„ì˜ í™•ë¥  ë¶„í¬ì™€ ì„ë² ë”©ëœ ê³µê°„ì˜ KL divergence(ë‘ ê³µê°„ì˜ ê²°í•©í™•ë¥ ê³¼ ì„ë² ë”© ê³µê°„ì˜ í™•ë¥ ë¶„í¬ì°¨ì´)ë¥¼ ì¤„ì´ë„ë¡ í•™ìŠµë©ë‹ˆë‹¤.

PCAê°€ ì„ í˜•ë¶€ë¶„ê³µê°„ì— íˆ¬ì˜ì‹œí‚¤ëŠ” ë°©ë²•ì¸ë° ë°˜í•´ t-SNEëŠ” ë¹„ì„ í˜•ì ì¸ ë¶€ë¶„ê³µê°„ì„ ì°¾ìœ¼ë¯€ë¡œ ë³µì¡í•œ ë°ì´í„°ë¥¼ ì‹œê°í™”í•  ë•Œ ìœ ìš©í•˜ê²Œ ì‚¬ìš©ë  ìˆ˜ ìˆìŠµë‹ˆë‹¤. t-SNEê°€ ìƒëŒ€ì ìœ¼ë¡œ ë³µì¡í•œ êµ¬ì¡°ë¥¼ í•™ìŠµí•˜ê¸° ë•Œë¬¸ì— PCAì— ë¹„í•´ì„œ ê³„ì‚°ëŸ‰ì´ ë§ìŠµë‹ˆë‹¤.

````python
# ë°ì´í„° ë¶ˆëŸ¬ì˜¤ê¸°
digits = datasets.load_digits(n_class=6)
X, y = digits.data, digits.target
n_samples, n_features = X.shape
print('ë°ì´í„° ìˆ˜:', n_samples) # ë°ì´í„° ìˆ˜ í™•ì¸
print('ì°¨ì› ìˆ˜:', n_features) # ì°¨ì›ìˆ˜ í™•ì¸
````

#### ğŸ‘¨â€ğŸ’» ì½”ë“œ: plotì„ ë„ì™€ì£¼ëŠ” í•¨ìˆ˜ ì •ì˜

````python
# plot helper í•¨ìˆ˜ ì •ì˜
def plot_embedding(X, title):
    _, ax = plt.subplots()
    # ì •ê·œí™”
    X = MinMaxScaler().fit_transform(X)
    # ìƒ‰ê¹”ë¡œ ìˆ«ìë¡œ scatter í‘œì‹œ
    for digit in digits.target_names:
        ax.scatter(
            *X[y == digit].T,
            marker=f"${digit}$",
            s=60,
            color=plt.cm.Dark2(digit),
            alpha=0.425,
            zorder=2,
        )
    # ì´ë¯¸ì§€ ê·¸ë¦¼ í‘œì‹œ
    shown_images = np.array([[1.0, 1.0]])
    for i in range(X.shape[0]):
        # ëª¨ë“  ìˆ«ì ì„ë² ë”©ì„ scatterí•˜ê³ , ìˆ«ì ê·¸ë£¹ì— annotation boxë¥¼ ë³´ê¸°
        dist = np.sum((X[i] - shown_images) ** 2, 1)
        # ë³´ê¸° ì‰½ê²Œ í•˜ê¸° ìœ„í•´ ë„ˆë¬´ ê°€ê¹Œìš´ ë°ì´í„°ëŠ” ë³´ì—¬ì£¼ì§€ ì•Šê¸°
        if np.min(dist) < 4e-3:
            continue
        # ì´ë¯¸ì§€ í•©ì¹˜ê¸°
        shown_images = np.concatenate([shown_images, [X[i]]], axis=0)
        imagebox = offsetbox.AnnotationBbox(
            offsetbox.OffsetImage(digits.images[i], cmap=plt.cm.gray_r), X[i]
        )
        imagebox.set(zorder=1)
        ax.add_artist(imagebox)

    ax.set_title(title)
    ax.axis("off")
````

t-SNEë¥¼ ì´ìš©í•˜ì—¬ ë†’ì€ ì°¨ì›ì˜ ìˆ«ì ë°ì´í„°ë¥¼ 2ì°¨ì›ì— ì‹œê°í™”í•©ë‹ˆë‹¤.

````python
# t-SNE ì ìš©
transformer = TSNE(n_components=2, random_state=0)
projection = transformer.fit_transform(X, y)

# t-SNE ê²°ê³¼ ì‹œê°í™”
plot_embedding(projection, 't-SNE embedding')
plt.show()
````

ìœ„ ì‹œê°í™” ê·¸ë¦¼ì„ ë³´ë©´, ê° ë°ì´í„°ê°€ ìˆ«ì í´ë˜ìŠ¤ì— ì˜í•´ì„œ êµ¬ë¶„ëœ ê²ƒì„ í™•ì¸í•  ìˆ˜ ìˆëŠ”ë°, ë°ì´í„°ì˜ íŒ¨í„´ì„ ì˜ í•™ìŠµí•œ embedding ê³µê°„ì„ ì°¾ì•„ ì‹œê°í™”í•˜ì—¬ ì´ 2ì°¨ì› ê³µê°„ë§Œìœ¼ë¡œë„ ë°ì´í„°ê°€ ì„œë¡œ ë¹„ìŠ·í•˜ê³  ë‹¤ë¥¸ íŠ¹ì§•ì„ ì˜ ì°¾ì•˜ë‹¤ëŠ” ê²ƒì„ ì˜ë¯¸í•©ë‹ˆë‹¤. ì´ë¡œë¶€í„° ë¶„ë¥˜ ëª¨ë¸ ë“±ì„ ì´ìš©í•˜ì§€ ì•Šê³ ë„ ë°ì´í„°ì˜ êµ¬ë¶„ë˜ëŠ” íŠ¹ì§•ì„ íŒŒì•…í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.

ìœ„ì—ì„œ ì‚´í´ë³¸ Truncated SVDë¥¼ ì ìš©í•œ ê²ƒê³¼ ë¹„êµí•´ë´…ë‹ˆë‹¤.

````python
# Truncated SVD ì ìš©
transformer = TruncatedSVD(n_components=2)
projection = transformer.fit_transform(X, y)

# TruncatedSVD ê²°ê³¼ ì‹œê°í™”
plot_embedding(projection, 'TruncatedSVD embedding')
plt.show()
````

Truncated SVD ì‹œê°í™”ë¡œë¶€í„° ë°ì´í„°ë“¤ì´ ì–´ëŠì •ë„ êµ¬ë¶„ì€ í•  ìˆ˜ ìˆìœ¼ë‚˜ í´ë˜ìŠ¤ ë³„ ê±°ë¦¬ê°€ ì˜ ë³´ì´ì§€ ì•ŠëŠ” ê²ƒì„ í™•ì¸í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤. ì´ì²˜ëŸ¼ T-SNEëŠ” í´ë˜ìŠ¤ ë³„ êµ¬ë¶„ì„ ëª…í™•í•˜ê²Œ í•˜ëŠ” ë°ì— ê°•ì ì„ ê°€ì§„ ê²ƒì„ í™•ì¸í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.

# 데이터 결측치 확인

```python
# 데이터 내 결측치가 있는지 확인하기
data.isnull()
```

```python
# 데이터 내 NULL 값이 있는지 변수 별로 확인하기

data.isnull().sum()
```

```python
# 결측치를 하나라도 포함하는 행의 인덱스를 확인
smaple_indices_with_nan = data.isnull().sum(axis=1) > 0

#결측치를 하나라도 포함하는 행들을 확인
data[smaple_indices_with_nan]
```

```python
# 결측치를 "8개 이상" 포함하는 행의 인덱스를 확인
sample_indices_with_nan = data.isnull().sum(axis=1) >= 8
# 결측치를 "8개 이상" 포함하는 행들이 어떤것인지 확인
data_with_nans = data[sample_indices_with_nan]
print(len(data_with_nans))

data_with_nans
```

```python
# NA를 가진 feature들을 변수로 정의하여 확인
columns_with_nan = ['Country', 'Abbreviation','Gross tertiary education enrollment (%)', 'Population','Unemployment rate','Urban_population','Latitude','Longitude']

# 위에서 선택한 8개 feature들이, 가설대로 모두 NaN인지 확인
data_with_nans[columns_with_nan].isnull().all(axis=1).sum()
```

```python
# 음이 아닌 값을 가지는 변수 내에 음수값이 있는지를 확인
# any() 를 사용하면 여러 값들에 대해서 하나라도 있는지 확인
(data['subscribers'] < 0).any()
```

```python
# 위도와 경도 각 변수의 최소값과 최대값 확인
print(data[['Latitude','Longitude']].min())
print(data[['Latitude','Longitude']].max())
```

```python
# 데이터 요약 통계량 보기
data.describe()
```

# 데이터 분포 살펴보기

```python
# 데이터 시각화를 위한 패키지 임포트
import matplotlib.pyplot as plt
import seaborn as sns
```

## 데이터 이상치 확인
> 데이터에 포함된 이상치를 파악해보기 위해 BoxPlot 시각화를 사용

```python
# boxplot을 확인할 변수에 다른 피쳐를 지정하면 다른 피쳐에도 boxplot을 그릴 수 있다.
target_feature = 'lowest_monthly_earnings'

# boxplot 보기
sns.boxplot(data,y=target_feature)
plt.show()
```

```python
# 이상치 제거를 위한 분위수 확인
Q3 = data['lowest_monthly_earnings'].quantile(0.75) # 75% 분위수
Q1 = data['lowest_monthly_earnings'].quantile(0.25) # 25% 분위수
IQR = Q3 - Q1 # 사분위수 범위
print(f"Q1: {Q1}")
print(f"Q3: {Q3}")
print(f"IQR: {IQR}")
```

```python
# 히스토 그램 보기
sns.histplot(data,x=target_feature) # kde 옵션은 막대 뿐만 아니라 밀도 형태의 선으로 보여주는것
plt.show()
```

```python
# 꼬리가 긴 분포를 가진 변수의 히스토그램을 로그스케일로 확인
sns.histplot(data,x=target_feature, log_scale=(False,True))
plt.show()
```

```python
# 막대그래프의 막대로 범주를 확인할 변수

category_feature = 'category'

#막대그래프의 높이로 값을 확인할 변수
target_feature = 'lowest_monthly_earnings'

# 막대그래프 생성
barplot = sns.barplot(data,x=category_feature, y= target_feature, color='C0', errorbar=None)

# matplotlib에서 x축 레이블의 위치, 방향 설정변경
loc, labels = plt.xticks()
barplot.set_xticklabels(labels, rotation=90)

plt.title('Lowest Monthly Earnings by Category')
plt.show()
```

## 변수간 관계 확인

```python
# 데이터의 상관계수 행렬을 생성
corr = data.corr(nummeric_only=True)  # 데이터셋으로부터 각 변수별 상관계수를 계산하여 행렬형태로 나타내는 함수.
# 피어슨 상관계수를 이용
# numberic_only는 숫자형 변수만 상관계수를 계산하도록 하는 옵션
# figuree에서 생략될 부분을 지정하는 mask 행렬을 생성 - 실제로는 mask 없이 다 생성해도 괜찮음
mask = np.ones_like(corr, dtype=bool)
mask = np.triu(mask)

# 시각화될 그림의 크기를 지정
# 히트맵 형태로 상관행렬 시각화하기
plt.figure(figsize=(13,10))
sns.heatmap(data=corr, annot=True, fmt='.2f', mask,linewidths=.5, cmap='rdYIBu_r')
plt.title('Correlation Matrix')
plt.show()
```
